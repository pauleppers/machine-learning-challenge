```python
import pandas as pd
import joblib  # for saving model
```

# Read the .CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()

```

## Select features (columns)

```python
# Set features. This will also be used as your x values.
X = df.drop(['koi_disposition'
             ,'koi_srad','koi_slogg', 'koi_insol_err2','koi_slogg_err1'
            ,'koi_prad_err2','koi_tce_plnt_num','koi_teq','koi_insol','koi_impact_err2'
            ,'koi_depth','koi_srad_err2','koi_kepmag','koi_period_err1','koi_duration_err1'
            , 'koi_time0bk_err1', 'koi_period_err2','koi_steff', 'koi_prad_err1', 'koi_impact_err1'
            , 'koi_depth_err2', 'koi_slogg_err2', 'koi_depth_err1', 'koi_duration_err2', 'koi_duration'
             ], axis = 1)   # 'koi_steff_err1', 'koi_time0bk_err2', 'koi_period', 'koi_srad_err1','dec'
y = df['koi_disposition']
```

##  Apply One-Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Label encode data set to 0, 1 or 2
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from tensorflow.keras.utils import to_categorical
one_hot_y = to_categorical(encoded_y)
one_hot_y
label_encoder.classes_
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y, random_state=42)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale X to ensure equal weight
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

# Transform the training and testing data using X_scaler
X_trained_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_trained_scaled.shape
```

# Train the Model



```python
#Create a Deep Learning Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense  # same # neurons, neuron gets signal from previous layer

model = Sequential()
model.add(Dense(units=30, input_dim=X_trained_scaled.shape[1], activation='relu'))  # relu for regression >=0 output
model.add(Dense(units=y_train.shape[1], activation='softmax')) # output probablity 0-1
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
             metrics=['accuracy'])

# Fit the model to the training data
model.fit(
    X_trained_scaled,
    y_train,
    epochs=1000,
    shuffle=True,
    verbose=0
)

# Quantify our Trained Model
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=0)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
# 40/1000 Normal Neural Network - Loss: 0.5122777699605824, Accuracy: 0.8890160322189331
# 30/1000 Loss: 0.3615263437242748, Accuracy: 0.8913043737411499 Loss: 0.3717369191024614, Accuracy: 0.8867276906967163
# 50/1000 Loss: 0.29635227613099924, Accuracy: 0.8735697865486145
# *beginning Scaler 30/1000 Normal Neural Network - Loss: 0.3550397614592546, Accuracy: 0.8667048215866089
# beginning MinMax Normal Neural Network - Loss: 0.33912580311980345, Accuracy: 0.8907322883605957
# above with all features Normal Neural Network - Loss: 0.2822408057595827, Accuracy: 0.8913043737411499
# final Normal Neural Network - Loss: 0.33505188661800095, Accuracy: 0.8524027466773987
```

# Deep Learning

```python
# Define the Model
deep_model = Sequential()
deep_model.add(Dense(units=30, input_dim=X_trained_scaled.shape[1], activation='relu'))
deep_model.add(Dense(units=30, activation='relu'))
deep_model.add(Dense(units=y_train.shape[1], activation='softmax'))
deep_model.summary()
```

```python
# Compile the model
deep_model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
             metrics=['accuracy'])

# Fit the model to the training data
deep_model.fit(X_trained_scaled, y_train, epochs=400, shuffle=True, verbose=2)
```

```python
model_loss_deep, model_accuracy_deep = deep_model.evaluate(
    X_test_scaled, y_test, verbose=0)
print(f"Deep Neural Network - Loss: {model_loss_deep}, Accuracy: {model_accuracy_deep}")
# minmax Deep Neural Network - Loss: 0.3108148748050838, Accuracy: 0.8535469174385071 
# units = 50, epoch = 500 Loss: 0.5037802626939333, Accuracy: 0.8724256157875061
# units = 30 epoch = 500 Deep Neural Network - Loss: 0.3703496129480052, Accuracy: 0.8821510076522827
# 30-30   500 Deep Neural Network - Loss: 0.5805560660034895, Accuracy: 0.8581235408782959
# 30-30 1000 Deep Neural Network - Loss: 1.0756981359193745, Accuracy: 0.8295194506645203
#30-30 400 Deep Neural Network - Loss: 0.33063479719227457, Accuracy: 0.8872997760772705
# above with all features Deep Neural Network - Loss: 0.274038371014104, Accuracy: 0.8890160322189331
# epoch 300,, 16 features, Deep Neural Network - Loss: 0.33465538724609045, Accuracy: 0.8689931631088257
```
