```python
import pandas as pd
from sklearn import tree
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
X = df.drop(['koi_disposition','koi_srad','koi_slogg', 'koi_insol_err2','koi_slogg_err1'
            ,'koi_prad_err2','koi_tce_plnt_num','koi_teq','koi_insol','koi_impact_err2'
            ,'koi_depth','koi_srad_err2','koi_kepmag','koi_period_err1','koi_duration_err1'
            , 'koi_time0bk_err1', 'koi_period_err2','koi_steff', 'koi_prad_err1', 'koi_impact_err1'
            , 'koi_depth_err2', 'koi_slogg_err2', 'koi_depth_err1', 'koi_duration_err2', 'koi_duration'
            , 'koi_steff_err1', 'koi_time0bk_err2', 'koi_period', 'koi_srad_err1','dec'
             ,'koi_fpflag_ec'], axis = 1)

y = df['koi_disposition']
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train
```

```python
# Single Tree
planet = tree.DecisionTreeClassifier()  #model this is one tree not a forrest
planet = planet.fit(X_train,y_train)   # train - class
print(planet.score(X_test,y_test))    # score
planet.feature_importances_
planet_df = pd.DataFrame(sorted(zip(planet.feature_importances_,X.columns), reverse=True))
planet_df
# score with all 0.8438215102974829
# score with top 14 is 0.8415331807780321
# score with top 11 is 0.8426773455377574
# score with top 10 is 0.8409610983981693 appears to start loosing accuracy once less than 10
```
