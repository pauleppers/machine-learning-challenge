# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()

# X = df.drop('koi_disposition', axis=1)
X = df.drop(['koi_disposition','koi_srad','koi_slogg', 'koi_insol_err2','koi_slogg_err1'
            ,'koi_prad_err2','koi_tce_plnt_num','koi_teq','koi_insol','koi_impact_err2'
            ,'koi_depth','koi_srad_err2','koi_kepmag','koi_period_err1','koi_duration_err1'
            , 'koi_time0bk_err1', 'koi_period_err2','koi_steff', 'koi_prad_err1', 'koi_impact_err1'
            , 'koi_depth_err2', 'koi_slogg_err2', 'koi_depth_err1', 'koi_duration_err2', 'koi_duration'
            , 'koi_steff_err1', 'koi_time0bk_err2'
             ,], axis = 1)
y = df['koi_disposition']
```

##  Apply One-Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Label encode data set to 0, 1 or 2
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from tensorflow.keras.utils import to_categorical
one_hot_y = to_categorical(encoded_y)
one_hot_y
label_encoder.classes_
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
# X = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y, random_state=42)
```

```python
X_train[:1]
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
# Scale X to ensure equal weight
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)

# Transform the training and testing data using X_scaler
X_trained_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_trained_scaled.shape
```

# Train the Model



```python
#Create a Deep Learning Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense  # same # neurons, neuron gets signal from previous layer

model = Sequential()
model.add(Dense(units=50, input_dim=X_trained_scaled.shape[1], activation='relu'))  # relu for regression >=0 output
model.add(Dense(units=3, activation='softmax')) # output probablity 0-1
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
             metrics=['accuracy'])

# Fit the model to the training data
model.fit(
    X_trained_scaled,
    y_train,
    epochs=60,
    shuffle=True,
    verbose=0
)

# model.predict([X_test_scaled])
# Quantify our Trained Model
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=0)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
#epoch = 600 Normal Neural Network - Loss: 0.7714941046398892, Accuracy: 0.8775743842124939
# reduced features, epoch = 100 Loss 0.3421185939344716, Accuracy: 0.8346682190895081
#reduced, epoch = 300   Loss: 0.36308987332972564, Accuracy: 0.8403890132904053
# reduced, epoch = 60 Loss: 0.3574535254755609, Accuracy: 0.8443936109542847
```

# Deep Learning

```python
# Define the Model
deep_model = Sequential()
deep_model.add(Dense(units=50, input_dim=X_trained_scaled.shape[1], activation='relu'))
deep_model.add(Dense(units=50, activation='relu'))
deep_model.add(Dense(units=y_train.shape[1], activation='softmax'))
deep_model.summary()
```

```python
# Compile the model
deep_model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
             metrics=['accuracy'])

# Fit the model to the training data
deep_model.fit(X_trained_scaled, y_train, epochs=100, shuffle=True, verbose=2)
```

```python
model_loss_deep, model_accuracy_deep = deep_model.evaluate(
    X_test_scaled, y_test, verbose=0)
print(f"Deep Neural Network - Loss: {model_loss_deep}, Accuracy: {model_accuracy_deep}")
```
